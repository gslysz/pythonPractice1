```python
def do_compare_old_vs_new(filename):
    import re
    basePath1 = "C:\\Gord\\output\\simpleDelta\\Original\\"
    basePath2 = "C:\\Gord\\output\\simpleDelta\\AfterCompression\\"
    graphsBasePath = "C:\\Gord\\output\\simpleDelta\\Graphs\\"

    p=re.compile('.*(simpleDelta_P3_00_\d+_)(\d+)')

    filename1 = os.path.basename(filename)
    print("filename1= "+ filename1)
    
    filepath2 = filename

    m = p.match(filename)
    if not m:
        print("Problem with the filename - ", filename)
        return

    
    filename1 = m.group(1) + str(int(m.group(2)) + 12).rjust(2,'0') + ".txt"
    filepath1 = basePath1 + filename1

    print("filepath1= " + filepath1)
    

    if not os.path.exists(filepath1):
        print("Filepath1 doesn't exist: " , filepath1)
        return
    
    graphfilepath =graphsBasePath + re.sub("txt","png",filename1)
    print("graphfilepath= "+ graphfilepath)

    data1 = pd.read_csv(filepath1)
    data2 = pd.read_csv(filepath2)

    fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(14,6))

    axes[0].plot(data1)
    axes[0].set_title('Original')
    #axes[0].set_ylim(-0.000045, 0.00002)

    ylim = axes[0].get_ylim()
    
    axes[1].plot(data2)
    axes[1].set_title('SimpleDelta')
    axes[1].set_ylim(ylim[0],ylim[1])

    plt.savefig(graphfilepath)
    plt.close()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import re

with os.scandir("C:\\Gord\\output\\simpleDelta\\AfterCompression") as files:
    for file in files:
        if file.name.endswith(".txt") and file.name.startswith("simpleDelta_P3_00_0") and file.is_file():
            do_compare_old_vs_new(file.path)
```
